```python
# This is a comment. The computer ignores it. We use it for notes.
# Let's put the number 10 into a box called 'age'
age = 10

# Let's put some text into a box called 'name'
# Text in Python is called a 'string' and needs quotes.
name = "Uche"

# A list of fruits
fruits = ["apple", "banana", "orange"]
print(fruits)
print(fruits[1]) # We can grab the first item (Python starts counting at 0!)

# A dictionary for a student
student = {
"name": "Amina",
"age": 22,
"course": "Data Science"
}
print(student)
print(student["name"]) # We grab data using its label, or 'key'

# First, we need to 'import' our tools from the toolbox.
import pandas as pd
# Let's read a CSV file from a URL (I've provided a simple one for us)
url = "data/simple_student_data_2.csv"
df = pd.read_csv(url)
# Now, let's take a peek at our data
print("First 5 rows:")
print(df.head())
print("\nQuick statistics:")
print(df.describe())



# Let's assume our DataFrame has a 'score' column
# We can filter the entire DataFrame like this
high_scorers = df[df['score'] > 90]
print(high_scorers)

# Import the visualization libraries
import matplotlib.pyplot as plt
import seaborn as sns

# Set a style to make our charts look nice
sns.set_style("whitegrid")
# Let's create a bar plot using Seaborn
plt.figure(figsize=(8,5)) # Set the size of our chart
sns.barplot(data=df, x='subject', y='score')
plt.title("Average Score by Subject") # Add a title
plt.xlabel("Subject") # Label the x-axis
plt.ylabel("Average Score") # Label the y-axis
plt.show() # This displays the chart


# A histogram shows us how the scores are spread out
plt.figure(figsize=(8,5))
sns.histplot(data=df, x='score', bins=10)
plt.title("Distribution of Student Scores")
plt.show()
```
